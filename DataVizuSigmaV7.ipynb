# Correction $\sigma$

As sigma depends on the pressure P(v), to be more precise in its calculus, we need to correct this value.

Correction sigma1

$P_{up}$ know by experimentation = 0.22 bar

$$ \sigma_{up} = \frac{Pv(T) - P_{up} }{0.5*\rho*C_{up}^2}  $$

Equation $Pv(T)$

$log P = A - \frac{B}{T + C}$

p in mm Hg has to be convert in Pa by this operation: /750 * 101325

https://onlinelibrary.wiley.com/doi/pdf/10.1002/9781118135341.app1

```python
A = 7.94917
B = 1657.462
C = 227.02
rho = 998.3 # [kg/m^3]
Pup = 0.22* 100000  + 101325 # [bar] to [Pa]


df1['Pv(T) [Pa]'] = ( 10 ** (A - B/(df1['Tw [°C]'] + C)) )/750 * 101325

df1['sigma_corr [-]'] = -1 * (df1['Pv(T) [Pa]'] - (df1['M1 [bar rel] ']*100000 + 101325) ) / (0.5*rho*(df1['Cup [m/s]'])**2)   
#utilisé la pression M1, et -1 * car sigma étaient négatifs
```

```python
df1['Cd [-]'] = (2* df1['Fd [N]'] ) / (rho * (df1['Cup [m/s]'])**2 * S_up)
df1['Cl [-]'] = (2* df1['Fl [N]'] ) / (rho * (df1['Cup [m/s]'])**2 * S_up)


df2['Cd [-]'] = (2* df2['Fd [N]'] ) / (rho * (df2['Cup [m/s]'])**2 * S_up)
df2['Cl [-]'] = (2* df2['Fl [N]'] ) / (rho * (df2['Cup [m/s]'])**2 * S_up)

df3['Cd [-]'] = (2* df3['Fd [N]'] ) / (rho * (df3['Cup [m/s]'])**2 * S_up)
df3['Cl [-]'] = (2* df3['Fl [N]'] ) / (rho * (df3['Cup [m/s]'])**2 * S_up)

df4['Cd [-]'] = (2* df4['Fd [N]'] ) / (rho * (df4['Cup [m/s]'])**2 * S_up)
df4['Cl [-]'] = (2* df4['Fl [N]'] ) / (rho * (df4['Cup [m/s]'])**2 * S_up)

df5['Cd [-]'] = (2* df5['Fd [N]'] ) / (rho * (df5['Cup [m/s]'])**2 * S_up)
df5['Cl [-]'] = (2* df5['Fl [N]'] ) / (rho * (df5['Cup [m/s]'])**2 * S_up)
```
